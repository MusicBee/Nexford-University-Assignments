```python
# Function to get employee details from dictionary of lists
def get_employee_details(name):

    # Convert DataFrame into dictionary of lists
    salary_data = df.to_dict(orient="list")

    try:
        for i in range(len(salary_data["EmployeeName"])):
            if salary_data["EmployeeName"][i].lower() == name.lower():
                emp_name = salary_data["EmployeeName"][i]
                gender = salary_data["JobTitle"][i]
                salary = salary_data["TotalPayBenefits"][i]
                Benfit = salary_data["Benefits"][i]

                # Apply rules
                if 10000 < salary < 20000:
                    level = "A1"
                if 7500 < salary < 30000 and gender.lower() == "female":
                    level = "A5-F"


                return {
                    "EMPLOYEE_NAME": emp_name,
                    "JOB_TITLE": gender,
                    "SALARY": salary,
                    "BENEFITS": Benfit
                }


        return f"Employee '{name}' not found."

    # Exception and Error handling to address potential issues.
    except Exception as e:
        return f"Error occurred: {e}"
```

```python
# Test My Function with a sample employee's name as input to returns their details.
print(get_employee_details("MICHAEL MORRIS"))
```

```python
# Function to export employee details into CSV and zip it
def export_employee_profile(name):

    #Get employee details
    details = get_employee_details(name)


    #Exception Returns no employee found when its empty
    if not details:
        print(f"Employee '{name}' not found. No file created.")
        return

    # Create a DataFrame from employee details
    df = pd.DataFrame([details])


    # Define folder and file names
    folder_name = "Employee Profile"
    os.makedirs(folder_name, exist_ok=True)  # Create folder if not exists
    csv_file = os.path.join(folder_name, f"{name}_profile.csv")


    # Save employee details to CSV
    df.to_csv(csv_file, index=False)
    print(f"Employee details saved as {csv_file}")


    # Create a ZIP file containing the CSV
    zip_filename = f"{folder_name}.zip"
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(csv_file, arcname=f"{name}_profile.csv")


    # Print when folder has been created
    print(f"Employee profile zipped into {zip_filename}")
```

```python
export_employee_profile("NATHANIEL FORD")
```
